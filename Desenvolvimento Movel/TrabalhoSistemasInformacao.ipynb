# 1. O que é o Pandas e qual é a sua principal funcionalidade?

R: Pandas nada mais é que uma biblioteca para Python, que fornece estruturas de dados, ferramentas para manipulação e análises de dados de maneira eficiente e rápida, bastante utilizada em ciência de dados. Algumas coisas possíveis a se fazer com Pandas é Estruturação de dados, Carregamentos de dados, Seleção de Dados, Agrupamento de Dados, Limpeza de dados e Visualização dos dados. 

Duas estruturas importantes no Pandas é a Series e o DataFrame.
Series: é uma matriz unidimensional que pode conter qualquer tipo de dado, números inteiros, floats, strings e objetos Python. Ela também possui índice, que permite a localização e acesso rápido a valores específicos da série.
DataFrame: Conjunto de dados tabulares, no qual cada coluna tem um tipo de dado representado por índice e coluna, parece bastante com uma planilha do Excel. No qual cada coluna pode ser de um tipo de dado diferente.

1- Como está estruturado o dataset?

```python
structure = df_acidentes.shape

print(structure)
print(structure[0], 'linhas')
print(structure[1], 'colunas')
```

2- Quais são as colunas presentes no dataset?

```python
df_acidentes.columns.value_counts().to_frame('').reset_index().rename(columns={'index' : 'Colunas'})
```

3- Quais os tipos das colunas do dataset?

```python
df_acidentes.dtypes.to_frame('Tipo').reset_index().rename(columns={'index' : 'Colunas'})
```

4- Qual a quantidade de valores nulos por coluna (em ordem decrescente)?

```python
df_acidentes.isna().sum().sort_values(ascending=False).to_frame('Nulos').reset_index().rename(columns={'index' : 'Colunas'})
```

5- Qual a quantidade de ocorrências por tipo de acidente?

```python
df_acidentes['classificacao_acidente'].value_counts() \
  .to_frame('Total') \
  .reset_index().rename(columns={'index': 'Classificação'}) \
  .to_csv('resumo_classificacao.csv', index=False, sep=';')

pd.read_csv('resumo_classificacao.csv', sep=';')
```

6- Quais fases do dia apresentam mais ocorrências por classificação de gravidade do acidente? (top-3 por classificação)

```python
df_acidentes[['fase_dia','classificacao_acidente']] \
  .value_counts().to_frame('total').reset_index() \
  .sort_values(by=['fase_dia', 'total'], ascending=[False, False], ignore_index=True) \
  .groupby(by=['classificacao_acidente']).head(4)
```

7- Quais munícipios apresentam mais ocorrências relacionadas a ingestão de álcool pelo condutor? (top-10)

```python
df_acidentes.query(expr='causa_acidente in ["Ingestão de álcool pelo condutor"]') \
  .groupby(by=['municipio']) \
  .agg({'municipio': ['count'], 'horario': ['min', 'max']}) \
  .sort_values(by=[('municipio', 'count')], ascending=[False]).head(10)
```

8- Quais são as condições metereológicas que apresentam menos ocorrências? (top-5)

```python
df_acidentes.value_counts(subset=['condicao_metereologica']).to_frame('total').tail(5) \
  .reset_index().sort_values(by=[('total')],
               ascending=[True],
               ignore_index=True)
```
