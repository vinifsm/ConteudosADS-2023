```python
def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument('--verbose')
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument('--window-size=1920, 1200')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument('--remote-allow-origins=*')
    options.add_argument(f'user-agent={user_agent}') # 403 Forbidden
    # options.add_argument('--lang=pt');
    # options.add_argument('--accept-lang=pt');
    # options.add_experimental_option('prefs', {'intl.accept_languages': 'pt,pt_BR'})
    driver = webdriver.Chrome(options=options)
    return driver

def open_page(url, delay=10):
  driver = web_driver()
  driver.get(url)
  time.sleep(delay)
  res = driver.page_source.encode('utf-8')
  soup = BeautifulSoup(res, 'html.parser')
  return driver, soup

def get_soup(driver):
  res = driver.page_source.encode('utf-8')
  soup = BeautifulSoup(res, 'html.parser')
  return  soup

# Beautiful Soup (Pretty-Print)
def pp(element):
  if element is not None and hasattr(element, 'prettifyx'):
    try:
      print(element.prettify())
    except:
      pass
```

```python
lista = []

table = soup.find('table', attrs={'class': 'dbi'})
tr = table.find_all('tr')[3:25]

for items in tr:
  bd = items.find('th', attrs={'class': 'pad-l'}).find('a').contents[0].text
  bd_link = items.find('th', attrs={'class': 'pad-l'}).find('a')['href']
  tipo = items.find('th', attrs={'class': 'small pad-r'}).text
  if 'Multi-model' in tipo:
    tipo = items.find('th', attrs={'class': 'pad-r'}).find('span', attrs={'class': 'infobox infobox_r'}).text
  lista.append({
      'Banco': bd,
      'Link': bd_link,
      'Tipo': tipo.split(',')
  })

df = pd.DataFrame(lista)

df
```

```python
lista = []

table = soup.find('table', attrs={'class': 'dbi'})
tr = table.find_all('tr')[3:]

for items in tr:
    bd = items.find('th', attrs={'class': 'pad-l'}).find('a').contents[0].text
    bd_link = items.find('th', attrs={'class': 'pad-l'}).find('a')['href']
    driver.get(bd_link)
    soup2 = get_soup(driver)

    for tr in soup2.find_all('tr'):
      if tr.find('td').text.strip() == 'Server operating systems':
        sos = ' '.join(list(map(lambda x: x.text, tr.find_all('td')[1].contents))).split('  ')

    lista.append({
      'Banco': bd,
      'Link': bd_link,
      'Sistemas Compat√≠veis': sos,
      'Quantidade de Sistemas': len(sos)
    })

df = pd.DataFrame(lista)
df.sort_values('Quantidade de Sistemas', ascending=False)
```
